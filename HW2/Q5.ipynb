```python
import numpy as np
```

```python
def epsilon_greedy(Q,s):  ### epsilon greedy selection Epsilon = 0.1
    pr = np.random.choice([0,1], p = [0.1,0.9])
    if(pr == 0):
        a = np.random.randint(3)
    else:
        a = np.argmax(Q[s,:])
    return a
```

```python
Q = np.random.rand(3,3)  ## Q table
r = np.array([1,2,3])
temp = np.empty(Q.shape)
err = 0.00001
gama = 0.5
alpha = 0.7
noe = 1000 ##Number of epochs
```

```python
for i in range(0,noe):
    s = np.random.randint(3)
    a = epsilon_greedy(Q,s)
    sn = a
    t = 0
    while(1):
        t+=1
        mx = np.max(Q[sn,:])
        temp[:] = Q[:] 
        Q[s,a] = Q[s,a] + alpha*(r[a] + gama*mx - Q[s,a])
        if(np.sum(np.abs(Q-temp))<err):
            break
        s = sn
        a = epsilon_greedy(Q,s)
        sn = a
    print("i =", i, "Consecutive time stepError", np.sum(np.abs(Q-temp)))
```

```python
pi = np.empty(3)
pi[0] = np.argmax(Q[0,:])
pi[1] = np.argmax(Q[1,:])
pi[2] = np.argmax(Q[2,:])
## Policy
print("Pi_1 (Action to choose when on state 1) =", pi[0]+1)
print("Pi_2 (Action to choose when on state 1) =", pi[1]+1)
print("Pi_3 (Action to choose when on state 1) =", pi[2]+1)
```
